```python
t = 1, 3, 5 # 팩킹 실행 (1, 3, 5)
print(t) 

print(type(t))

l = list(t)
print(l, type(l))

x, y, z = t # 언팩킹 실행
print(x, y, z)


a, b = [3, 6] # (3, 6)변경 후 언팩킹 발생
print(a, b)
```

```python
t = 1, 3, 5  # 패킹
print(t)

# 언패킹
x, y, z = t
print(x)
print(y)
print(z)

# 변수와 항목(요소) 수 불일치 Error
a, b = t
```

```python
x = (1, 3)
print(x)

y = list(x)
print(y)

z = tuple(y)
print(z)
```

```python
a = (1, 2)
b = (3, 4)

print(a + b)  # (1, 2, 3, 4)
```

```python
print((1, 2) * 3)  # (1, 2, 1, 2, 1, 2)
```

```python
t = (1, 2, 3)

print(2 in t)  # True
print(5 in t)  # False
```

```python
t = (10, 20, 30, 40)

print(t[1])    # 20
print(t[1:3])  # (20, 30)
```

```python
t = (1, 2, 2, 3)

print(t.count(2))  # 2의 개수 -> 2
print(t.index(2))  # 2의 첫 위치 -> 1
```

```python
t = (1, 2, 3)

lst = list(t)    # 튜플 -> 리스트
lst.append(4)

t2 = tuple(lst)  # 리스트 -> 튜플
print(t2)        # (1, 2, 3, 4)
```
