```python
# append() : 맨 뒤에 1개 추가
fruits = ["apple", "banana"]
fruits.append("orange")
print(fruits)
```

```python
# insert() : 원하는 위치에 추가
fruits = ["apple", "banana"]
fruits.insert(1, "orange")
print(fruits)
```

```python
# extend() : 여러 개 추가
nums = [1, 2, 3]
nums.extend([4, 5, 6])
print(nums)
```

```python
# remove() : 값으로 삭제
fruits = ["apple", "banana", "orange"]
fruits.remove("banana")
print(fruits)

# 수정은 인덱스 할당, 슬라이싱 할당으로 수정
```

```python
# pop() : 인덱스로 삭제 후 반환
nums = [10, 20, 30]
result = nums.pop() # 마지막 인덱스 삭제 후 반환
print(result)
print(nums)

nums = [10, 20, 30]
result = nums.pop(1) # 인덱스1 삭제 후 반환
print(result)
print(nums)
```

```python
# clear() : 전체 삭제
nums = [10, 20, 30]
nums.clear()
print(nums)
```

```python
# del 문
nums = [10, 20, 30, 40]
del nums[1]
print(nums)
```

```python
# reverse() : 순서 뒤집기, 정렬과 상관없이 현재 순서를 거꾸로 만든다.
nums = [10, 20, 30, 40]
nums.reverse()
print(nums)
```

```python
# count() : 특정 값이 몇개인지
nums = [10, 20, 10, 30, 10]
print(nums.count(10)) # 3
```

```python
# index() : 특정 값의 인덱스
fruits = ["apple", "banana", "orange"]
print(fruits.index("banana")) # 1
```

```python
# 정렬 메서드
nums = [50, 10, 40, 20]
nums.sort() # 오름차순
print(nums)

nums = [50, 10, 40, 20]
nums.sort(reverse=True) # 내림차순
print(nums)
```

```python
nums = [50, 10, 40, 20]
result = sorted(nums)
print(nums)
print(result)

nums = [50, 10, 40, 20]
result = sorted(nums, reverse=True)
print(result)

fruits = ["banana", "apple", "orange"]
result = sorted(fruits)
print(result)

# key매갲변수를 이용하여 문자열 길이로 정렬
fruits = ["watermelon", "kiwi", "banana", "apple"]
result = sorted(fruits, key=len)
print(result)

# 매개변수 key값으로는 함수가 온다. 뒤에서 람다함수를 배우면
# key값으로 람다식을 대입하여 다양한 요구 사항의 정렬도 가능하다.
# 자기 자신을 정렬할때는 sort()메서드, 다양한 형태의 정렬된 리스트를 반환할때는 sorted() 내장함수 사용
```
