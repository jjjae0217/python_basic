```python
greet = "Hello"
while greet: # 문자열의 경우 공백이면 False, 아니면 True
    print(greet)
    greet = greet[1:]
```

```python
a, b = 0, 10 
# 다중 할당
# 내부적으로 오른쪽의 튜플 (0, 10)을 왼쪽의 변수들에 언패킹해서 각각 저장

while a < b:
    print(a)
    a = a + 1 #a += 1
```

```python
# break가 없는 경우

for i in range(3):
    print(i)
else:
    print("정상 종료")
```

```python
# break가 있는 경우

for i in range(3):
    if i == 1:
        break
    print(i)
else:
    print("정상 종료") # break가 실행되었으므로 else는 실행되지 않는다.
```
