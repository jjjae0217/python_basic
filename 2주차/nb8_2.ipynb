```python
try:
    num = int(input("숫자 입력 : "))
    result = 10 / num
    print(result)

except ValueError:
    print("숫자를 입력하세요.")

except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
```

```python
try:
    print(10 / 2)
except:
    print("오류 발생")
finally:
    print("프로그램 종료")
```

```python
try:
    num = int(input("숫자 입력: "))
    result = 100 / num

except ValueError:
    print("숫자를 입력해야 합니다.")

except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")

else:
    print("계산 결과:", result)

finally:
    print("프로그램 종료")
```

```python
try:
    with open("test.txt", "r", encoding="utf-8") as f:
        text = f.read()

except FileNotFoundError:
    print("파일이 없습니다.")

else:
    print(text)

finally:
    print("파일 읽기 작업 종료")
```
