```python
# 문제 1. 결측치 확인 : 구매금액이 결측치(None)인 데이터의 개수를 구하시오.
# 1. 이상치 처리
for order in orders:

    amount = order["구매금액"]

    if amount is not None:

        if amount < 0 or amount > 1000000:
            print(order)
```

```python
# 문제 2. 결측치 처리 : 결측치를 제외한 구매금액 평균을 계산한 후 결측치를 평균값으로 대체하시오.
for o in orders:
    if o["구매금액"] is not None:
        if o["구매금액"] > 1000000:
            o["구매금액"] = 1000000

        if o["구매금액"] < 0:
            o["구매금액"] = 0

for order in orders:
    print(orders)
```

```python
# 문제 3. 이상치 탐지 : 구매금액이 (0 ~ 1,000,000원) 범위를 벗어나면 이상치로 판단한다.
# 2. 결측치 처리(None --> 평균)
count = 0

for o in orders:
    if o["구매금액"] is None:
        count = count + 1

print(count)
```

```python
# 문제 4. 이상치 처리 :이상치는 다음 기준으로 수정하시오.
total = 0
count = 0

for o in orders:
    if o["구매금액"] is not None:
        total = total + o["구매금액"]
        count = count + 1

avg = int(round(total / count, 1))
print(avg)

for o in orders:
    if o["구매금액"] is None:
        o["구매금액"] = avg

for order in orders:
    print(orders)
```

```python
# 문제 5. 회원 가입 연수 계산 (특성 추출) : 현재 연도를 기준으로 회원 가입 후 몇 년이 지났는지 계산하여 "가입연수" 컬럼을 추가하시오.

# 3. 가입연수
year = 2026

for o in orders:
    join_year = int(o["가입일"][0:4])
    o["가입연수"] = year - join_year

for order in orders:
    print(orders)
```
