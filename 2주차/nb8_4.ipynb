```python
CSV 읽기
      │
      ▼
DictReader
      │
      ▼
list(reader)
      │
      ▼
결측치 처리
      │
      ▼
이상치 처리
      │
      ▼
파생변수 생성
(평균점수)
      │
      ▼
특성추출
(합격 여부 또는 등급)
      │
      ▼
기술통계
(mean, median, min, max 등)
```

```python
#1 CSV 읽기
import csv

with open("score.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    data = list(reader)

for student in data:
    print(student)
```

```python
# 2. 결측치 처리
kor_scores = []
eng_scores = []
ages = []

for student in data:

    if student["국어"] not in ["", "결측", "NaN"]:
        kor_scores.append(int(student["국어"]))

    if student["영어"] not in ["", "결측", "NaN"]:
        eng_scores.append(int(student["영어"]))

    if student["나이"] not in ["", "결측", "NaN"]:
        ages.append(int(student["나이"]))

avg_kor = sum(kor_scores) / len(kor_scores)
avg_eng = sum(eng_scores) / len(eng_scores)
avg_age = sum(ages) / len(ages)

for student in data:

    if student["국어"] in ["", "결측", "NaN"]:
        student["국어"] = round(avg_kor)

    else:
        student["국어"] = int(student["국어"])

    if student["영어"] in ["", "결측", "NaN"]:
        student["영어"] = round(avg_eng)

    else:
        student["영어"] = int(student["영어"])

    if student["나이"] in ["", "결측", "NaN"]:
        student["나이"] = round(avg_age)

    else:
        student["나이"] = int(student["나이"])

for student in data:
    print(student)
```

```python
# 3. 이상치 처리
for student in data:

    if student["나이"] < 0 or student["나이"] > 100:
        student["나이"] = round(avg_age)

    if student["국어"] < 0 or student["국어"] > 100:
        student["국어"] = round(avg_kor)

    if student["영어"] < 0 or student["영어"] > 100:
        student["영어"] = round(avg_eng)
for student in data:
    print(student)
```

```python
# 4. 파생변수 생성 (총점)
for student in data:
    student["총점"] = student["국어"] + student["영어"]

for student in data:
    print(student)
```

```python
# 5. 파생변수 생성 (평균)
for student in data:
    student["평균"] = round(student["총점"] / 2, 1)

for student in data:
    print(student)
```

```python
# 6. 특성추출 (합격여부)
for student in data:

    if student["평균"] >= 60:
        student["합격여부"] = "합격"
    else:
        student["합격여부"] = "불합격"

for student in data:
    print(student)
```

```python
# 7. 순위 계산
for current_student in data:

    rank = 1

    for other_student in data:

        if other_student["총점"] > current_student["총점"]:
            rank += 1

    current_student["순위"] = rank

for student in data:
    print(student)
```

```python
# 8. 평균(mean)
scores = []

for student in data:
    scores.append(student["평균"])

mean_score = sum(scores) / len(scores)

print(mean_score)
```

```python
# 9. 중앙값(median)
scores = []

for student in data:
    scores.append(student["평균"])

scores.sort()

n = len(scores)

if n % 2 == 1:
    median = scores[n // 2]
else:
    median = (scores[n // 2 - 1] + scores[n // 2]) / 2

print(median)
```

```python
# 10. 최댓값(max)
scores = []

for student in data:
    scores.append(student["평균"])

print(max(scores))
```

```python
# 11. 최솟값(min)
scores = []

for student in data:
    scores.append(student["평균"])

print(min(scores))
```

```python
# 12. 결과 CSV 저장
import csv

with open("score_result.csv",
          "w",
          newline="",
          encoding="utf-8-sig") as f:

    writer = csv.DictWriter(
        f,
        fieldnames=data[0].keys()
    )

    writer.writeheader()
    writer.writerows(data)

print("저장 완료")
```
