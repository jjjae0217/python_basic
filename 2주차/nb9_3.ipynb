```python
def hello():
    print("안녕하세요!")

def execute(func):
    func() # 함수호출연산자를 사용했다는 건 func매개변수값이 함수다

execute(hello)
```

```python
# 일반 함수 정의
def add1(a, b):
    return a + b

# 람다 함수
add2 = lambda a, b : a + b # add2에는 "lambda a, b : a + b" 람다가 반환한 함수(객체)가 대입

r1 = add1(5, 1)
print(r1)

r2 = add2(1, 3)
print(r2)
```

```python
s = [('홍길동', 90), ('이순신', 85), ('강감찬', 100)]
r = sorted(s, key=lambda x: x[1], reverse=True)
# lambda x: x[1] 
# x(리스트의 요소 튜플)를 입력받아 x[1](튜플의 인덱스1값)를 반환한다.
print(r)
# [('강감찬', 100), ('홍길동', 90), ('이순신', 85)] 점수의 내림차순으로 정렬됨

① 원본 리스트를 하나씩 꺼낸다.
('홍길동', 90)
('이순신', 85)
('강감찬', 100)

② key 함수의 인수로 하나씩(리스트 사이즈 만큼 key함수를 반복 호출하면서 하나씩 전달)  전달한다.

③ sorted 내장함수는 인수로 입력받은 key함수의  반환값(x[1] 요소)으로 정렬한다.
```

```python
def score(item): # 아이템을 입력받아 1번째 요소를 반환 ex)  ('홍길동', 90)가 인수로 입력되면 90 반환
    return item[1]

s = [
    ('홍길동', 90),
    ('이순신', 85),
    ('강감찬', 100)
]

r = sorted(s, key=score, reverse=True)

print(r)
```

```python
students = [
    {"이름": "홍길동", "나이": 20, "점수": 90},
    {"이름": "이순신", "나이": 18, "점수": 85},
    {"이름": "김유신", "나이": 22, "점수": 95}
]

# "나이" 값을 기준으로 오름차순 정렬
result = sorted(students, key=lambda x: x["나이"])
print(result)

# x : 리스트의 각 딕셔너리
# x["나이"] : 각 딕셔너리의 "나이" 값을 꺼냄
# sorted()는 이 값을 기준으로 정렬한다.
```

```python
# "점수"를 기준으로 내림차순 정렬
result = sorted(students, key=lambda x: x["점수"], reverse=True)
print(result)
```

```python
# "이름"을 기준으로 오름차순 정렬
result = sorted(students, key=lambda x: x["이름"])
print(result)
```

```python
students = [
    {"이름": "홍길동", "반": 2, "점수": 90},
    {"이름": "이순신", "반": 1, "점수": 90},
    {"이름": "김유신", "반": 1, "점수": 95}
]

# 1순위: 반, 2순위: 점수
result = sorted(students, key=lambda x: (x["반"], x["점수"]))
print(result)
```

```python
names = ["banana", "apple", "kiwi"]

# 문자열 길이가 가장 짧은 요소 반환
result = min(names, key=lambda x: len(x))

print(result) # kiwi
```

```python
names = ["banana", "apple", "kiwi"]

# 문자열 길이가 가장 긴 요소 반환
result = max(names, key=lambda x: len(x))

print(result) # banana
```

```python
nums = [1, 3, 5, 8]

# 각 요소가 짝수인지(True/False)로 변환
# 하나라도 True이면 True 반환

result = any(map(lambda x: x % 2 == 0, nums))

print(result) # True
```

```python
nums = [1, 3, 5, 8]

# 각 요소가 짝수인지(True/False)로 변환
# 하나라도 True이면 True 반환

r1 = map(lambda x: x % 2 == 0, nums)
print("r1: ", r1)
r2 = list(r1)
print("r2: ", r2)

result = any(map(lambda x: x % 2 == 0, nums))

print(result) # True
```

```python
nums = [2, 4, 6, 8]

# 각 요소가 짝수인지(True/False)로 변환
# 모두 True이면 True 반환
result = all(map(lambda x: x % 2 == 0, nums))

print(result) #True
```
