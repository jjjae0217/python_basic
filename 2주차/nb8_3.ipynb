```python
with 자원해지가 필요한 객체 as 변수:
    실행할 코드
```

```python
# with문 사용하지 않는 경우
f = open("students.csv", "r", encoding="utf-8")
print(f.read())
# print(10 / 0)
f.close()   # 직접 닫아야 함. 실수로 빼먹거나, 실행전 에러가 발생하면 메모리 누수가 발생한다. 

# try-except-finally문으로 자원 해지
try:
    f = open("students.csv", "r", encoding="utf-8")
    print(f.read())
    print(10 / 0)

except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")

finally:
    f.close()
    print("자원해지 성공")

# with문을 사용하는 경우
with open("students.csv", "r", encoding="utf-8") as f:
    print(f.read())
```

```python
with open("students.txt", "r", encoding="utf-8") as f:
    # 한 줄씩 읽기.
    for line in f: 
        print(line, end="") # 라인 끝의 줄바꿈이 문자 포함되어 두번 줄바꿈이 일어난다(print함수도 줄바꿈이 기본값)
```

```python
with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        print(line.strip()) # line에 양쪽 공백제거(줄바꿈포함)하여 print함수의 기본 줄바꿈만으로 출력
```

```python
# 한 라인을 구분자","를 이용하여 두개의 문자열 값으로 저장
with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, score = line.strip().split(",")
        print(name, int(score))
```

```python
# 딕셔너리 만들기
students = {}

with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, score = line.strip().split(",")
        students[name] = int(score)

print(students)
```

```python
names = []
scores = []
with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, score = line.strip().split(",")
        # print(name, int(score))
        names.append(name)
        scores.append(score)


print(names)
print(scores)
```

```python
s = [] # [(), (), ()]

with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        a = line.strip().split(",")
        # print(name, int(score))
        
        # print(a)
        s.append(tuple(a))

print(s)
```

```python
import csv

with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)

    data = list(reader)

print(data)

data[2]["영어"] = 80

print(data[2])

data[2]["영어"] = 'None' #csv 원본 파일은 변화하지 않는다.

print(data[2])
```

```python
import csv

# students_result.csv 파일을 쓰기(w) 모드로 열기
# newline="" : 빈 줄이 추가되는 현상 방지
# encoding="utf-8-sig" : 한글이 엑셀에서 깨지지 않도록 저장
with open("students_result.csv", "w", newline="", encoding="utf-8-sig") as f:
    # data[0].keys() → 첫 번째 딕셔너리의 키(컬럼명) 사용    
    writer = csv.DictWriter(f, fieldnames=data[0].keys())
    # 헤더(컬럼명) 저장
    writer.writeheader()   
    # data 리스트에 있는 모든 딕셔너리를 한 번에 저장
    writer.writerows(data)  

print("CSV 저장 완료")
```
