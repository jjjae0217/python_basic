```python
# 제곱값 만들기
squares = []

for i in range(1, 6):
    squares.append(i ** 2)

print(squares)
```

```python
evens = [i for i in range(1, 11) if i % 2 == 0]
print(evens)
```

```python
names = ["kim", "lee", "park"]
result = [name.upper() for name in names]
print(result)
```

```python
students = [
    {"이름": "김철수", "성적": 95},
    {"이름": "이영희", "성적": 88},
    {"이름": "박민수", "성적": 77}
]

names = [s["이름"] for s in students]
print(names)

scores = [s["성적"] for s in students]
print(scores)

# 우수학생 추출
good_students = [
    s["이름"]
    for s in students
    if s["성적"] >= 90
]
print(good_students)
```

```python
squares = {
    i: i**2
    for i in range(1, 6)
}

print(squares)
```

```python
# 이름과 길이
names = ["홍길동", "이순신", "김유신"]
result = {name: len(name) for name in names}
print(result) # {'홍길동': 3, '이순신': 3, '김유신': 3}
```

```python
accounts = [
    "admin-1234",
    "guest-0000",
    "user-7777"
]

result = {a.split("-")[0]: a.split("-")[1] for a in accounts}

print(result)
```

```python
accounts = [
    "admin-1234",
    "guest-0000",
    "user-7777"
]

result = {
    user: password for user, password in (account.split("-") for account in accounts)
} # accounts 결과 [(,),(,),..]를 다시 컴프리헨션

print(result)
```

```python
scores = [95, 80, 67, 100]

result = {
    s: "합격" if s >= 80 else "불합격"
    for s in scores
}

print(result)
```

```python
#키-값 반전
scores = {
    "홍길동": 90,
    "이순신": 80,
    "김유신": 100
}

result = {value: key for key, value in scores.items()}

print(result)
```
