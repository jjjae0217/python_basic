```python
a = {1, 2, 3}
b = {3, 4, 5}


#교집합
print(a & b)
print(a. intersection(b))

#합집합
print(a | b)   #{1,2,3,4,5}
print(a.union(b))

#차집합
print(a - b)   #{1,2}
print(a.difference(b))
```
