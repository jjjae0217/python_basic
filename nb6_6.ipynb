```python
# 문제 1. 결측치 확인 : 구매금액이 결측치(None)인 데이터의 개수를 구하시오.

missing_count = 0

for order in orders:
    if order["구매금액"] is None:
        missing_count += 1

print("결측치 개수:", missing_count)
```

```python
# 문제 2. 결측치 처리 : 결측치를 제외한 구매금액 평균을 계산한 후 결측치를 평균값으로 대체하시오.
amounts = []

for order in orders:
    if order["구매금액"] is not None:
        amounts.append(order["구매금액"])

avg = sum(amounts) / len(amounts)

for order in orders:
    if order["구매금액"] is None:
        order["구매금액"] = avg

for order in orders:
    print(order)
```

```python
# 문제 3. 이상치 탐지 : 구매금액이 (0 ~ 1,000,000원) 범위를 벗어나면 이상치로 판단한다.
for order in orders:

    amount = order["구매금액"]

    if amount < 0 or amount > 1000000:
        print(order)
```

```python
# 문제 4. 이상치 처리 :이상치는 다음 기준으로 수정하시오.
for order in orders:

    if order["구매금액"] > 1000000:
       order["구매금액"] = 1000000

    elif order["구매금액"] < 0:
         order["구매금액"] = 0

for order in orders:
    print(orders)
```

```python
# 문제 5. 회원 가입 연수 계산 (특성 추출) : 현재 연도를 기준으로 회원 가입 후 몇 년이 지났는지 계산하여 "가입연수" 컬럼을 추가하시오.

from datetime import datetime

current_year = datetime.now().year

for order in orders:

    join_year = int(order["가입일"][:4])

    order["가입연수"] = current_year - join_year

for order in orders:
    print(orders)
```
