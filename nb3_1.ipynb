```python
# 1. 정수 (int)
a = 10
print(a)
print(type(a))
```

```python
# 2. 실수 (float) -> 부등 소수점 수(float) : 오차를 가진다 - 최대 16자리
b = 3.14
print(b)
print(type(b))

print(0.1 + 0.2)
```

```python
# 3. 복소수 (complex)
c = 3 + 4j
print(c)
print(type(c))
```

```python
# 4. 논리값 (bool)
d = True
print(d)
print(type(d))
```

```python
# 5. 문자열 (str)
e = "Python"
print(e)
print(type(e))

print(e[1])

e[1] = "K" # error 문자열객체 불변이다
```

```python
# 6. 리스트 (list)
f = [1, 2, 3]
print(f)
print(type(f))

print(f[1])

f[1] = "abc" 
print(f)
```

```python
# 7. 튜플 (tuple)
g = (1, 2, 3)
print(g)
print(type(g))

print(g[1])

g[1] = 9
```

```python
# 8. 딕셔너리 (dict)
h = {"id": "admin", "pw":"1234"} #"(1,3,5,7)":"1234
print(h)
print(type(h))

print(h["id"])

h["pw"] = "0000"
print(h)
```

```python
# 9. 집합 (set)
i = {1, 2, 3}
print(i)
print(type(i))

#print(i[0])
```

```python
# 10. 범위 (range) #

j = range(0, 5)
print(j)
print(type(j))
print(list(j))
```

```python
# 11. 바이트 (bytes)
k = b"abc"
print(k)
print(type(k))
```

```python
# 12. 바이트배열 (bytearray)
l = bytearray(b"abc")
print(l)
print(type(l))
```

```python
# 13. 없음 (NoneType)
m = None
print(m)
print(type(m))
```

```python
# 14. 함수 (function)
def hello():
    print("Hello")

print(hello)
print(type(hello))
```

```python
# 15. 모듈 (module)
import math

print(math)
print(type(math))
```

```python
# 16. 클래스 (type)
class Person:
    pass

print(Person)
print(type(Person))
```
