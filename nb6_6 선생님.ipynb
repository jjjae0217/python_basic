```python
# 문제 11 가장 구매금액이 높은 고객 TOP 3를 출력하시오. 11-2) 연평균 구매금액(소수점1자리ㅠ반올림)이 높은 구매고객 TOP3?#9. Top3

result = [] #구매금액으로만 만들어진 리스트 -> 리스트를 내림차순 정렬 후 위에서 3개만

for o in orders:
    result.append(o['구매금액'])

print(result)

result2 = sorted(result)
print(result2)


# bubble sort (단순 버전)
for i in range(len(result)):
    for j in range(len(result)):
        if result[i]["구매금액"] > result[j]["구매금액"]:
            temp = result[i]
            result[i] = result[j]
            result[j] = temp

print(result[0])
print(result[1])
print(result[2])
```
