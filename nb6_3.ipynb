```python
# Shallow Copy(얕은 복사)는 "1단계만 복사"하고 내부 객체의 주소(참조)는 그대로 복사한다. 
# 그래서 중첩된 리스트·딕셔너리 등은 공동 참조가 발생한다.

a = [[1, 2], [3, 4]]
b = a.copy()

# 바깥 리스트는 다른 객체
print(a is b)      # False

# 내부 리스트는 같은 객체
print(a[0] is b[0]) # True
```

```python
a = [[1, 2], [3, 4]]

b = a.copy()

b[0][0] = 999 # 내부리스트는 공동참조

print(a) # [[999, 2], [3, 4]]
print(b) # [[999, 2], [3, 4]]

b[1] = 7 # 바깥 리스트는 다른 객체
print(a) # [[999, 2], [3, 4]]
print(b) # [[999, 2], 7]
```

```python
import copy

a = [[1, 2], [3, 4]]

b = copy.deepcopy(a)

b[0][0] = 999

print(a)
print(b)
```
